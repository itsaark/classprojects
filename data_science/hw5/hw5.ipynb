```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10,7)})
df = pd.read_csv('winemag-data_first150k.csv')
df.head()
```

```python
sns.distplot(df['points'],bins=100,kde=False,color='red')
```

```python
plt.figure(figsize=(17,10))
ax = sns.countplot(x='country',data=df[df['country'].coun],palette='Set3')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
ca_df = df[df['country'] == 'Canada']
ca_df.head()
```

```python
plt.figure(figsize=(10,7))
ax = sns.boxplot(x='province',y='points',data=ca_df,palette=sns.color_palette("husl", 8))
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x="price", y="points", data=ca_df, kind="kde",color='red')
```

```python
top = df['country'].value_counts().nlargest(5)
```

```python
top_df = df[(df['country'] == 'US') | (df['country'] == 'Spain') | (df['country'] == 'Italy') | (df['country'] == 'Chile') | (df['country'] == 'France')]
top_df.head()
```

```python
g = sns.FacetGrid(top_df, hue="country",palette='Set2',size=10)
g.map(plt.scatter, "points", "price", alpha=.7)
g.add_legend()
```

```python
sns.violinplot(x="country", y="points", data=top_df)
```
